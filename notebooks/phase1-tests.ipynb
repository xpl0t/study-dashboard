```python
# JSON speichern & landen nativ

from dataclasses import dataclass, asdict
import json
  
@dataclass
class Semester:
    nummer: int
  
@dataclass
class Studiengang:
    name: str
    semester: list[Semester]
  
def speichere_studiengang(studiengang: Studiengang, dateiname: str) -> None:
    with open(dateiname, "w", encoding="utf-8") as f:
        json.dump(asdict(studiengang), f, ensure_ascii=False, indent=2)
  
def lade_studiengang(dateiname: str) -> Studiengang:
    with open(dateiname, "r", encoding="utf-8") as f:
        daten = json.load(f)
    return Studiengang(**daten)
  
if __name__ == "__main__":
    original = Studiengang("Angewandte künstliche Intelligenz", [ Semester(1) ])
    speichere_studiengang(original, "studium.json")
  
    geladen = lade_studiengang("studium.json")
    print(geladen)
```

```python
# JSON speichern & landen mit dataclasses_json
# %pip install dataclasses_json

from dataclasses import dataclass
from dataclasses_json import dataclass_json
  
@dataclass
class Semester:
    nummer: int

@dataclass_json
@dataclass
class Studiengang:
    name: str
    semester: list[Semester]
  
def speichere_studiengang(studiengang: Studiengang, dateiname: str) -> None:
    with open(dateiname, "w", encoding="utf-8") as f:
        f.write(studiengang.to_json())
  
def lade_studiengang(dateiname: str) -> Studiengang:
    with open(dateiname, "r", encoding="utf-8") as f:
        str = f.read()
        return Studiengang.from_json(str)

if __name__ == "__main__":
    original = Studiengang("Angewandte künstliche Intelligenz", [ Semester(1) ])
    speichere_studiengang(original, "studium.json")
  
    geladen = lade_studiengang("studium.json")
    print(geladen)
```
